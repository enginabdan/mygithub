```python
# Relation "Flight_Mean" with "Exams_Mean","Sprt_Mean"

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,4), sharey=False, sharex=False)

ax[0].plot(stu.Flight_Mean, stu.Exams_Mean, "r+")
ax[0].set_xlabel("Flight_Mean")
ax[0].set_ylabel("Exams_Mean")
ax[0].set_xlim([50,100])
ax[0].set_ylim([80,100])
ax[0].axvline(x=stu.Flight_Mean.mean(), ls="--", lw=2, color="blue")
ax[0].axhline(y=stu.Exams_Mean.mean(), ls="--", lw=2, color="blue")

ax[1].plot(stu.Flight_Mean, stu.Sprt_Mean, "c.")
ax[1].set_xlabel("Flight_Mean")
ax[1].set_ylabel("Sprt_Mean")
ax[1].set_xlim([50,100])
ax[1].set_ylim([75,100])
ax[1].axvline(x=stu.Flight_Mean.mean(), ls="--", lw=2, color="red")
ax[1].axhline(y=stu.Sprt_Mean.mean(), ls="--", lw=2, color="red")

plt.tight_layout();
```

```python
# Relation "Flight_Mean" with Multiple Intelligences one by one, hue="Plt_Or_Not", size="Gender"

fig, ax = plt.subplots(nrows=2, ncols=4, figsize=(20,12), sharey=False, sharex=False)
a = range(2)
b = range(4)
k = 7

while k < 15:
    for i in a:
        for j in b:        
            sns.scatterplot(data=stu,
                            x=stu.Flight_Mean,
                            y=stu[stu.columns[k]],
                            hue="Plt_Or_Not",
                            size="Gender",
                            sizes=(100,10),
                            ax=ax[i,j],
                            palette="muted")
            ax[i,j].set_xlabel("Flight_Mean")
            ax[i,j].set_ylabel(stu.columns[k])
            ax[i,j].set_xlim([50,100])
            ax[i,j].set_ylim([5,35])
            ax[i,j].axvline(x=stu.Flight_Mean.mean(), ls="--", lw=2, color="white")
            ax[i,j].axhline(y=stu[stu.columns[k]].mean(), ls="--", lw=2, color="white")
            ax[i,j].legend(loc='upper left', title="Plt_Or_Not")
            k+=1

plt.tight_layout();
```

```python
# Relation "Exams_Mean" with Multiple Intelligences one by one, hue="Plt_Or_Not", size="Gender"

fig, ax = plt.subplots(nrows=2, ncols=4, figsize=(20,12), sharey=False, sharex=False)
a = range(2)
b = range(4)
k = 7

while k < 15:
    for i in a:
        for j in b:        
            sns.scatterplot(data=stu,
                            x=stu.Exams_Mean,
                            y=stu[stu.columns[k]],
                            hue="Plt_Or_Not",
                            size="Gender",
                            sizes=(100,10),
                            ax=ax[i,j],
                            palette="muted")
            ax[i,j].set_xlabel("Exams_Mean")
            ax[i,j].set_ylabel(stu.columns[k])
            ax[i,j].set_xlim([80,100])
            ax[i,j].set_ylim([5,35])
            ax[i,j].axvline(x=stu.Exams_Mean.mean(), ls="--", lw=2, color="white")
            ax[i,j].axhline(y=stu[stu.columns[k]].mean(), ls="--", lw=2, color="white")
            ax[i,j].legend(loc='upper left', title="Plt_Or_Not")
            k+=1

plt.tight_layout();
```

```python
# Relation "Sprt_Mean" with Multiple Intelligences one by one, hue="Plt_Or_Not", size="Gender"

fig, ax = plt.subplots(nrows=2, ncols=4, figsize=(20,12), sharey=False, sharex=False)
a = range(2)
b = range(4)
k = 7

while k < 15:
    for i in a:
        for j in b:        
            sns.scatterplot(data=stu,
                            x=stu.Sprt_Mean,
                            y=stu[stu.columns[k]],
                            hue="Plt_Or_Not",
                            size="Gender",
                            sizes=(100,10),
                            ax=ax[i,j],
                            palette="muted")
            ax[i,j].set_xlabel("Sprt_Mean")
            ax[i,j].set_ylabel(stu.columns[k])
            ax[i,j].set_xlim([70,100])
            ax[i,j].set_ylim([5,35])
            ax[i,j].axvline(x=stu.Sprt_Mean.mean(), ls="--", lw=2, color="white")
            ax[i,j].axhline(y=stu[stu.columns[k]].mean(), ls="--", lw=2, color="white")
            ax[i,j].legend(loc='upper left', title="Plt_Or_Not")
            k+=1

plt.tight_layout();
```

```python
# Relation "Flight_Mean","Exams_Mean","Sprt_Mean" with Multiple Intelligences one by one
# "boxplot", "swarmplot"

fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(20,16))
a = range(3)

for i in a:
        sns.boxplot(x=stu.MI_Cat, y=stu[stu.columns[i+4]], ax = ax[i], palette="muted", color="white")
        sns.swarmplot(x=stu.MI_Cat, y=stu[stu.columns[i+4]], ax = ax[i], color="white")
plt.tight_layout();
```

```python
# "Countplot", "MI_Cat", hue="Plt_Or_Not"

fig, ax = plt.subplots(figsize=(10,7))

ax = sns.countplot(data=stu, x=stu.MI_Cat, hue="Plt_Or_Not", palette="muted")

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.15, p.get_height()+0.1))
plt.tight_layout;
```

```python
# "Countplot", "MI_Cat", hue="Gender"

fig, ax = plt.subplots(figsize=(10,7))

ax = sns.countplot(data=stu, x=stu.MI_Cat, hue="Gender", palette="muted")

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.07, p.get_height()+0.1))
plt.tight_layout;
```

```python
# "barplot", "MI_Cat", with "Flight_Mean","Exams_Mean","Sprt_Mean", hue="Plt_Or_Not"

fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(20,20))
a = range(3)

for i in a:
        sns.barplot(x=stu.MI_Cat, y=stu[stu.columns[i+4]], data=stu, ax = ax[i], hue="Plt_Or_Not", palette="muted", ci="sd")
        ax[i].set_ylim([50,100])
        for p in ax[i].patches:
                ax[i].annotate(round((p.get_height())), (p.get_x()+0.22, p.get_height()+1.0))
plt.tight_layout();
```

```python
# "PairGrid", "Flight_Mean","Exams_Mean","Sprt_Mean", hue="Plt_Or_Not"
# map_diag  >>> histplot
# map_upper >>> scatterplot
# map_lower >>> kdeplot


g = sns.PairGrid(stu,
                hue='Plt_Or_Not',
                diag_sharey=False,
                height=3,
                aspect=1,
                vars=['Flight_Mean', 'Exams_Mean','Sprt_Mean'])
g.map_diag(sns.histplot, palette="muted") 
g.map_upper(sns.scatterplot, palette="muted")
g.map_lower(sns.kdeplot)
plt.tight_layout();
```

```python
# "FacetGrid", col='Plt_Or_Not', hue="Gender"
# map >>> histplot

g = sns.FacetGrid(stu, col='Plt_Or_Not', hue="Gender", height=5, aspect=1, palette="muted")
g.map(sns.histplot, "MI_Cat")
plt.tight_layout();
```

```python
# "Heatmap" for all variables

plt.figure(figsize=(10,6))
sns.heatmap(stu.corr(), annot=True, cmap="BrBG")
plt.tight_layout();
```

```python
# "Heatmap" for the target "Plt_Or_Not" variable
plt.figure(figsize=(3,9))
x = stu.corr()[["Plt_Or_Not"]].sort_values(by="Plt_Or_Not", ascending=False)
sns.heatmap(x, annot=True, cmap="BrBG")
plt.show()
```
