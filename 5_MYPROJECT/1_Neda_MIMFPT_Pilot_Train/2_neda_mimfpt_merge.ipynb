#### Creating Student, Insructor & Grades Lists via related columns

```python
a = ["Ling","Math","Spa","Mus","Nat","Body","Interp","Intrap"]
j = 0

for i in a:
    if a[j] == "Ling":
        Ling_Stu_List1 = list(stu.sort_values(by=i+"_Int", ascending=False).reset_index().iloc[lambda x: x.index % 2 == 0].loc[:,"Student"])
        Ling_Stu_List2 = list(stu.sort_values(by=i+"_Int", ascending=False).reset_index().iloc[lambda x: x.index % 2 != 0].loc[:,"Student"])
        Ling_Ins_Name = list(ins.sort_values(by=i+"_Int", ascending=False).reset_index().loc[:,"insructor"])
        Ling_Ins_Grd = list(ins.sort_values(by=i+"_Int", ascending=False).reset_index().loc[:,i+"_Int"])
        j+=1
    elif a[j] == "Math":
        Math_Stu_List1 = list(stu.sort_values(by=i+"_Int", ascending=False).reset_index().iloc[lambda x: x.index % 2 == 0].loc[:,"Student"])
        Math_Stu_List2 = list(stu.sort_values(by=i+"_Int", ascending=False).reset_index().iloc[lambda x: x.index % 2 != 0].loc[:,"Student"])
        Math_Ins_Name = list(ins.sort_values(by=i+"_Int", ascending=False).reset_index().loc[:,"insructor"])
        Math_Ins_Grd = list(ins.sort_values(by=i+"_Int", ascending=False).reset_index().loc[:,i+"_Int"])
        j+=1
    elif a[j] == "Spa":
        Spa_Stu_List1 = list(stu.sort_values(by=i+"_Int", ascending=False).reset_index().iloc[lambda x: x.index % 2 == 0].loc[:,"Student"])
        Spa_Stu_List2 = list(stu.sort_values(by=i+"_Int", ascending=False).reset_index().iloc[lambda x: x.index % 2 != 0].loc[:,"Student"])
        Spa_Ins_Name = list(ins.sort_values(by=i+"_Int", ascending=False).reset_index().loc[:,"insructor"])
        Spa_Ins_Grd = list(ins.sort_values(by=i+"_Int", ascending=False).reset_index().loc[:,i+"_Int"])
        j+=1
    elif a[j] == "Mus":
        Mus_Stu_List1 = list(stu.sort_values(by=i+"_Int", ascending=False).reset_index().iloc[lambda x: x.index % 2 == 0].loc[:,"Student"])
        Mus_Stu_List2 = list(stu.sort_values(by=i+"_Int", ascending=False).reset_index().iloc[lambda x: x.index % 2 != 0].loc[:,"Student"])
        Mus_Ins_Name = list(ins.sort_values(by=i+"_Int", ascending=False).reset_index().loc[:,"insructor"])
        Mus_Ins_Grd = list(ins.sort_values(by=i+"_Int", ascending=False).reset_index().loc[:,i+"_Int"])
        j+=1
    elif a[j] == "Nat":
        Nat_Stu_List1 = list(stu.sort_values(by=i+"_Int", ascending=False).reset_index().iloc[lambda x: x.index % 2 == 0].loc[:,"Student"])
        Nat_Stu_List2 = list(stu.sort_values(by=i+"_Int", ascending=False).reset_index().iloc[lambda x: x.index % 2 != 0].loc[:,"Student"])
        Nat_Ins_Name = list(ins.sort_values(by=i+"_Int", ascending=False).reset_index().loc[:,"insructor"])
        Nat_Ins_Grd = list(ins.sort_values(by=i+"_Int", ascending=False).reset_index().loc[:,i+"_Int"])
        j+=1
    elif a[j] == "Body":
        Body_Stu_List1 = list(stu.sort_values(by=i+"_Int", ascending=False).reset_index().iloc[lambda x: x.index % 2 == 0].loc[:,"Student"])
        Body_Stu_List2 = list(stu.sort_values(by=i+"_Int", ascending=False).reset_index().iloc[lambda x: x.index % 2 != 0].loc[:,"Student"])
        Body_Ins_Name = list(ins.sort_values(by=i+"_Int", ascending=False).reset_index().loc[:,"insructor"])
        Body_Ins_Grd = list(ins.sort_values(by=i+"_Int", ascending=False).reset_index().loc[:,i+"_Int"])
        j+=1
    elif a[j] == "Interp":
        Interp_Stu_List1 = list(stu.sort_values(by=i+"_Int", ascending=False).reset_index().iloc[lambda x: x.index % 2 == 0].loc[:,"Student"])
        Interp_Stu_List2 = list(stu.sort_values(by=i+"_Int", ascending=False).reset_index().iloc[lambda x: x.index % 2 != 0].loc[:,"Student"])
        Interp_Ins_Name = list(ins.sort_values(by=i+"_Int", ascending=False).reset_index().loc[:,"insructor"])
        Interp_Ins_Grd = list(ins.sort_values(by=i+"_Int", ascending=False).reset_index().loc[:,i+"_Int"])
        j+=1
    elif a[j] == "Intrap":
        Intrap_Stu_List1 = list(stu.sort_values(by=i+"_Int", ascending=False).reset_index().iloc[lambda x: x.index % 2 == 0].loc[:,"Student"])
        Intrap_Stu_List2 = list(stu.sort_values(by=i+"_Int", ascending=False).reset_index().iloc[lambda x: x.index % 2 != 0].loc[:,"Student"])
        Intrap_Ins_Name = list(ins.sort_values(by=i+"_Int", ascending=False).reset_index().loc[:,"insructor"])
        Intrap_Ins_Grd = list(ins.sort_values(by=i+"_Int", ascending=False).reset_index().loc[:,i+"_Int"])
        j+=1
```
