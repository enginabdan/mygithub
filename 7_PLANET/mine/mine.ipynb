```python
start_time = datetime(1950, 1, 15, 0, 0, 0) # set the starting time
stop_time = datetime(2023, 4, 1, 0, 0, 0) # get the current time

target = ["JUPITER"]
observer = "SUN"
reference = "J2000"

planet = []
time = []
X = []
Y = []
Z = []
position = []

for i in target:
    start_time = start_time # reset the start time for each target
    while start_time < stop_time: # loop until current time is reached
        nowET = spice.str2et(str(start_time))
        Positions = spice.spkpos(i, nowET, reference, 'NONE', observer)
        planet.append(i.lower())
        time.append(start_time)
        X.append(Positions[0][0])
        Y.append(Positions[0][1])
        Z.append(Positions[0][2])
        position.append(Positions[1])
        start_time += timedelta(minutes=1) # add 1 minute to the current time
        if start_time >= stop_time:
            break # exit the loop if the current time has been reached
print(len(planet),": planet")
print(len(time),": time")
print(len(X),": X")
print(len(Y),": Y")
print(len(Z),": Z")
print(len(position),": position")
```

```python
df1 = pd.DataFrame({'source' : 'SUN',
                   'target': planet,
                   'time' : time,
                   'position_lt': position,
                   'X': X,
                   'Y': Y,
                   'Z': Z})
df1
```
